```python
signal = input("Enter traffic signal color: ")

if signal == "red":
    print("Stop")
elif signal == "green":
    print("Go")
elif signal == "yellow":
    print("Wait")
else:
    print("Invalid signal")
```

```python
battery = int(input("Enter phone battery percentage: "))

if battery >= 80 and battery <= 100:
    print("Battery Full")
elif battery >= 40 and battery < 80:
    print("Battery Medium")
elif battery >= 20 and battery < 40:
    print("Battery Low")
elif battery >= 0 and battery < 20:
    print("Battery Very Low - Please Charge")
else:
    print("Invalid battery percentage")
```

```python
usernames = ["admin", "kishore", "haneff"]
passwords = ["1234", "54321", "98765"]

u = input("Username: ")
p = input("Password: ")

if u in usernames:
    i = usernames.index(u)
    if p == passwords[i]:
        print("Account logged in")
    else:
        print("Wrong password")
else:
    print("Create an account")
```
